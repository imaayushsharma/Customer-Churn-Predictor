```python
import tensorflow as tf 
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
model=load_model("model.h5")
with open("label_encoder_gender.pkl","rb") as file:
    label_encoder_gender=pickle.load(file)
with open("ohe_encoder_geography.pkl","rb") as file:
    ohe_encoder_geography=pickle.load(file)
with open("scaler.pkl","rb") as file:
    scaler=pickle.load(file)
```

```python
input_data = {
    "CreditScore": 715,
    "Geography": "France",
    "Gender": "Male",
    "Age": 37,
    "Tenure": 5,
    "Balance": 84532.45,
    "NumOfProducts": 2,
    "HasCrCard": 1,
    "IsActiveMember": 0,
    "EstimatedSalary": 112450.30
}
input_df=pd.DataFrame([input_data])

```

```python
encoded_geo=ohe_encoder_geography.transform(input_df[["Geography"]])
```

```python
geo_df=pd.DataFrame(encoded_geo,columns=ohe_encoder_geography.get_feature_names_out(["Geography"]))
```

```python
geo_df
```

```python
input_df=pd.concat([input_df,geo_df],axis=1)
```

```python
input_df=input_df.drop(["Geography"],axis=1)
```

```python
input_df
```

```python
input_df["Gender"]=label_encoder_gender.transform(input_df["Gender"])
```

```python
input_df=scaler.transform(input_df)
```

```python
prediction=model.predict(input_df)
```

```python
prediiction_prob=prediction[0][0]
```

```python
if prediction < 0.5 :
     print("Not Exited")
else:
    print("Exited")
    
```
