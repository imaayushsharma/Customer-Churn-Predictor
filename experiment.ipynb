```python
## importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
## loading Dataset
data=pd.read_csv("Churn_Modelling.csv")
```

```python
## Checking Top 5 Records
data.head()
```

```python
## Checking Statistical Information
data.describe()
```

```python
## Removing unnecessary features
data.drop(["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
data.head()
```

```python
## Categorical to numeric - conversion
labelencoder_gender=LabelEncoder()
data["Gender"]=labelencoder_gender.fit_transform(data["Gender"])
```

```python
data.head()
```

```python
## Checking unique values present in gender column 
data["Gender"].unique()
```

```python
# Create encoder
ohe_geo = OneHotEncoder(sparse_output=False)

# Fit and transform
geo_encoded = ohe_geo.fit_transform(data[["Geography"]])

# Convert to DataFrame with proper column names
geo_df = pd.DataFrame(geo_encoded, columns=ohe_geo.get_feature_names_out(["Geography"]))

# (Optional) Concatenate with main dataset
data = pd.concat([data, geo_df], axis=1)
```

```python
data.head()
```

```python
## droping Geography column from df as there is no need of it now
data.drop(["Geography"],axis=1,inplace=True)
```

```python
## import pickle for save the model
import pickle
## pickle file
with open("label_encoder_gender.pkl","wb") as file:
    pickle.dump(labelencoder_gender,file)
```

```python
with open("ohe_encoder_geography.pkl","wb") as file:
    pickle.dump(ohe_geo,file)
```

```python
## seperate dependent and independent features
X = data.drop(["Exited"],axis=1)
Y = data.Exited
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.20)
```

```python
## converting Y_train/Y_test to array as tf not process over series 
Y_train = np.array(Y_train, dtype='float32')
Y_test = np.array(Y_test, dtype='float32')
```

```python
std_sclr=StandardScaler()
X_train=std_sclr.fit_transform(X_train)
X_test=std_sclr.transform(X_test)
```

```python
print(type(Y_train))
print(Y_train[:10])
print(Y_train.dtype)
```

```python
## importing ANN libraries
import tensorflow
from tensorflow.keras.models import Sequential ## ann sequential model
from tensorflow.keras.layers import Dense ## dense used for creating hidden neurons or hidden layer
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard # tensorboard for visualiztaion and early stopping to stop at early epochs if accuracy is const
import datetime

## creating Sequential model
model=Sequential([
    Dense(64,input_shape = (X_train.shape[1],),activation=('relu')), # for first neuron input_shape should be must in tuple
    Dense(32,activation=('relu')),
    Dense(20,activation=('relu')),
    Dense(1,activation=('sigmoid'))
])

## log_dir -> creates a date with time for every epochs
log_dir = "log_dir/" + datetime.datetime.now().strftime("%d-%m-%Y_%H-%M-%S")
```

```python
## manually choose optimizer and loss for compiling model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
opt = Adam(learning_rate=0.01)
```

```python
## Compiling model
model.compile(optimizer=opt,loss=BinaryCrossentropy(),metrics=["accuracy"])
```

```python
## Creating tensorboard callback
ensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
## Create Earlystopping Callback
early_stopping_callback = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

```python
## model fitting
history = model.fit(
    X_train, Y_train,
    validation_data=(X_test, Y_test),
    callbacks=[tensorboard_callback, early_stopping_callback],
    epochs=100
)
```

```python
## saving model in h5  as tf support h5 file formate
model.save('model.h5')
```
